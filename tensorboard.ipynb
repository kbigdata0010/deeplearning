```python
import pandas as pd

df = pd.read_csv('./datasets/ThoraricSurgery.csv', header=None)
df.info()
```

```python
X = df.drop(17, axis=1)
y = df[17]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import tensorflow as tf

np.random.seed(3)
tf.random.set_seed(3)

#네트워크 구조
model = Sequential()
model.add(Dense(30, input_dim =17 , activation='sigmoid' ))
model.add(Dense(1, activation='sigmoid' ))

model.compile(loss="binary_crossentropy" , optimizer='adam', metrics=['accuracy'])

model.fit(X , y , epochs=100 , batch_size=10)
```

```python
from tensorflow.keras.callbacks import TensorBoard
import datetime 
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import tensorflow as tf


log_dir = "logs/fit/"+ datetime.datetime.now().strftime("%Y%m%d-%H%M%S")

tensorboard_callbacks = TensorBoard(log_dir=log_dir , histogram_freq=1 )

np.random.seed(3)
tf.random.set_seed(3)

#네트워크 구조
model = Sequential()
model.add(Dense(30, input_dim =17 , activation='sigmoid' ))
model.add(Dense(1, activation='sigmoid' ))

model.compile(loss="binary_crossentropy" , optimizer='adam', metrics=['accuracy'])

model.fit(X , y , epochs=100 , batch_size=10 , callbacks=[tensorboard_callbacks])
```
